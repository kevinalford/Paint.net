```python
%load_ext autoreload
%autoreload 2

from PIL import Image
import numpy as np
import os
import random
import matplotlib.pyplot as plt
import pickle
```

```python
def get_pixel_arrays(directory, value):
    pixels = []
    #pixels.append(value)
    list_dir = os.listdir(directory)
    list_dir = [f.lower() for f in list_dir]
    list_dir = sorted(list_dir)
    for image in list_dir:
        print (image)
        image = "./" + directory + "/" + image
        im = Image.open(image, 'r')
        im = im.resize((128,128),Image.ANTIALIAS)
        pix_val = list(im.getdata())
        new_arr = np.expand_dims(pix_val, 0)
        new_arr = np.asarray(new_arr).squeeze()
        new_arr = new_arr.astype(float)# / 255
        
        corr_sized = np.zeros((128,128,3))
        new_arr = new_arr.T
        for i in range(3):
            corr_sized[:,:,i] = new_arr[i].reshape((128,128))
        pixels.append(corr_sized.astype(np.int32))
    return pixels
            
dali = get_pixel_arrays("dali", '0')

print(dali[1])
```
